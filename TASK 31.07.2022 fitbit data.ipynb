#8 . Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset 

```python
calories=df.groupby('Id')[['Calories']].mean()
```

```python
calories[calories['Calories']>2200]
```

#9. how many person are not a regular person with respect to activity try to find out those 

```python
regular=df.groupby('Id')[['ActivityDate']].count()
```

```python
regular[regular['ActivityDate']<20] #total activity days captured are 32 days assuming 20 days or more as regular 
```

#10 . who is the thired most active person in this dataset find out those in pandas and in sql both .

```python
df.groupby('Id')['Calories'].sum().sort_values(ascending=False).reset_index().iloc[[2]] # highest calories burnt in cumulative for entire data set is considered lowest 
```

#11 . who is the 5th most laziest person avilable in dataset find it out 

```python
df.groupby('Id')['Calories'].sum().sort_values().reset_index().iloc[[4]] # highest calories burnt in cumulative for entire data set is considered lowest 
```
